## Evaluation

Οι δύο κώδικες που παρουσιάζονται παρακάτω υλοποιούν ένα αυτοματοποιημένο framework αξιολόγησης για το αρχικό και το βελτιωμένο DS-STAR, χρησιμοποιώντας custom benchmark.

Το benchmark αποτελείται από μια σειρά από tasks, όπου για κάθε task ορίζονται το dataset, το ερώτημα, το αναμενόμενο πεδίο εξόδου (`expected_key`) και η αναμενόμενη απάντηση (`expected`). Τα expected keys προστέθηκαν ώστε ο evaluator να μπορεί να εντοπίζει στοχευμένα το πεδίο του JSON που περιέχει την τελική απάντηση και να πραγματοποιεί ακριβέστερη αξιολόγηση.

Η εκτέλεση πραγματοποιείται μέσω της συνάρτησης `run_dsstar()`, η οποία καλεί το `dsstar.py` μέσω `subprocess`, παρέχοντας το κατάλληλο dataset και το αντίστοιχο ερώτημα. Μετά την ολοκλήρωση της εκτέλεσης, ο κώδικας εντοπίζει το παραγόμενο αρχείο `result.json`, το φορτώνει και το μετατρέπει σε αντικείμενο Python για περαιτέρω επεξεργασία και αξιολόγηση.

Η αξιολόγηση πραγματοποιείται μέσω συναρτήσεων που υποστηρίζουν διαφορετικούς τύπους απαντήσεων, όπως αριθμούς, strings, λίστες και dictionaries. Επιπλέον, η διαδικασία αξιολόγησης περιλαμβάνει μηχανισμούς fallback, ώστε να μπορεί να ανακτήσει την απάντηση ακόμη και όταν αυτή δεν βρίσκεται ακριβώς στο αναμενόμενο JSON πεδίο.

Πριν την έναρξη της αξιολόγησης πραγματοποιείται έλεγχος εγκυρότητας του benchmark, ώστε να επιβεβαιωθεί ότι κάθε task περιέχει όλα τα απαραίτητα πεδία (`id`, `dataset`, `query`, `expected`, `expected_key`) και ότι τα αντίστοιχα datasets υπάρχουν στο σύστημα αρχείων.

Για κάθε task καταγράφονται πληροφορίες όπως η επιτυχής ή μη εκτέλεση, η ορθότητα της απάντησης και τα παραγόμενα αποτελέσματα. Επιπλέον, έχουν προστεθεί μηχανισμοί διαχείρισης σφαλμάτων και χρονικών ορίων (timeouts), ώστε η αποτυχία ενός task να μην διακόπτει ολόκληρη τη διαδικασία αξιολόγησης και σε μία τέτοια  περίπτωση σφάλματος, το task χαρακτηρίζεται ως αποτυχημένη εκτέλεση (`execution_success = False`) και η αξιολόγηση συνεχίζεται κανονικά με το επόμενο task.

Στο τέλος της διαδικασίας, τα αποτελέσματα αποθηκεύονται σε αρχεία JSON, και υπολογίζονται συνολικά metrics, όπως το ποσοστό επιτυχών εκτελέσεων (execution success) και το ποσοστό σωστών απαντήσεων (answer accuracy), τόσο συνολικά όσο και ανά κατηγορία ερωτήματος.

Οι δύο εκδόσεις ακολουθούν την ίδια βασική λογική αξιολόγησης, και η κύρια διαφορά τους είναι ότι η έκδοση που χρησιμοποιείται για το αρχικό DS-STAR δημιουργεί προσωρινό αρχείο ρυθμίσεων για κάθε task και αντιστοιχίζει μοναδικό `run_id` σε κάθε εκτέλεση, το οποίο δενγίνεται στην βελτιωμένη έκδοση του DS-STAR, όπου η διαχείριση των `run_id` πραγματοποιείται αυτόματα από το ίδιο το σύστημα


### Benchmark Runner - improved dsstar

```python
import json
import re
import csv
import subprocess
from pathlib import Path
from datetime import datetime

DSSTAR_DIR = Path("/content/DS_Star_Improvement_MM_projectc")
EVAL_DIR = Path("/content/custom_eval")
BENCHMARK_FILE = EVAL_DIR / "benchmark.json"
RESULTS_DIR = EVAL_DIR / "results"
RESULTS_DIR.mkdir(exist_ok=True)

MAX_TASKS = None


def run_dsstar(task):
    dataset_path = EVAL_DIR / task["dataset"]

    cmd = [
        "uv", "run", "python", "dsstar.py",
        "--data-files", str(dataset_path),
        "--query",
        task["query"] + f"\nReturn the final answer as valid JSON using exactly this top-level key: {task['expected_key']}.",
        "--max-rounds", "1"
    ]

    try:
        completed = subprocess.run(
            cmd,
            cwd=DSSTAR_DIR,
            capture_output=True,
            text=True,
            timeout=900
        )
    except subprocess.TimeoutExpired:
        return {
            "execution_success": False,
            "returncode": -1,
            "stdout": "",
            "stderr": "",
            "result_path": None,
            "result_json": None
        }

    stdout = completed.stdout
    stderr = completed.stderr

    match = re.search(
        r"OUTPUT:\s*(runs/.+?/final_output/result\.json)",
        stdout
    )

    result_json = None
    result_path = None

    if match:
        result_path = DSSTAR_DIR / match.group(1)

        if result_path.exists():
            try:
                with open(result_path, "r", encoding="utf-8") as f:
                    result_json = json.load(f)
            except json.JSONDecodeError:
                result_json = None

    return {
        "execution_success": result_json is not None,
        "returncode": completed.returncode,
        "stdout": stdout,
        "stderr": stderr,
        "result_path": str(result_path) if result_path else None,
        "result_json": result_json
    }

def flatten_json(obj):
    return json.dumps(obj, ensure_ascii=False).lower()


def extract_numbers(obj):
    text = flatten_json(obj)
    nums = re.findall(r"-?\d+(?:\.\d+)?(?:e-?\d+)?", text, flags=re.IGNORECASE)
    return [float(x) for x in nums]
def fuzzy_numeric_match(obj, expected, abs_tol=1e-3, rel_tol=0.01):
    nums = extract_numbers(obj)

    for n in nums:
        if compare_numbers(n, expected, abs_tol, rel_tol):
            return True

    return False

def get_expected_value(result_json, expected_key):
    if result_json is None:
        return None

    if isinstance(result_json, dict):
        if expected_key in result_json:
            return result_json[expected_key]

        if "final_answer" in result_json:
            final_answer = result_json["final_answer"]

            if isinstance(final_answer, dict) and expected_key in final_answer:
                return final_answer[expected_key]

            return final_answer

    return result_json


def compare_numbers(pred, expected, abs_tol=1e-3, rel_tol=0.01):
    try:
        pred = float(pred)
        expected = float(expected)
    except Exception:
        return False

    if abs(pred - expected) <= abs_tol:
        return True

    if expected != 0:
        return abs(pred - expected) / abs(expected) <= rel_tol

    return False

def compare_values(pred, expected):
    if isinstance(expected, (int, float)):
        return compare_numbers(pred, expected)

    if isinstance(expected, str):
        return str(expected).lower().strip() in flatten_json(pred)

    if isinstance(expected, list):
        text = flatten_json(pred)
        return all(str(item).lower() in text for item in expected)

    if isinstance(expected, dict):
        text = flatten_json(pred)

        for key, value in expected.items():
            key_ok = str(key).lower() in text

            if isinstance(value, (int, float)):
                value_ok = fuzzy_numeric_match(pred, value)
            else:
                value_ok = str(value).lower() in text

            if not (key_ok and value_ok):
                return False

        return True

    return False


def evaluate(result_json, expected_key, expected):
    if result_json is None:
        return False

    pred = get_expected_value(result_json, expected_key)

    if pred is not None and compare_values(pred, expected):
        return True

    return compare_values(result_json, expected)





def infer_category(task_id):
    if "basic" in task_id:
        return "dataset_understanding"
    if "_stat_" in task_id or "stat_" in task_id:
        return "statistical_reasoning"
    if "follow" in task_id:
        return "follow_up"
    if task_id.startswith("wrangle"):
        return "data_wrangling"
    if task_id.startswith("ml"):
        return "machine_learning"
    if task_id.startswith("multi"):
        return "multi_step"
    return "other"


def validate_benchmark(benchmark):
    errors = []

    for task in benchmark:
        if "id" not in task:
            errors.append("Task missing id")
            continue

        if "dataset" not in task:
            errors.append(f"{task['id']}: missing dataset")

        if "query" not in task:
            errors.append(f"{task['id']}: missing query")

        if "expected" not in task:
            errors.append(f"{task['id']}: missing expected")

        if "expected_key" not in task:
            errors.append(f"{task['id']}: missing expected_key")

        if "dataset" in task:
            dataset_path = EVAL_DIR / task["dataset"]
            if not dataset_path.exists():
                errors.append(f"{task['id']}: missing dataset {dataset_path}")

    if errors:
        raise ValueError("\n".join(errors))


def main():
    with open(BENCHMARK_FILE, "r", encoding="utf-8") as f:
        benchmark = json.load(f)

    validate_benchmark(benchmark)
    if MAX_TASKS is not None:
        benchmark = benchmark[:MAX_TASKS]
        #benchmark = benchmark[4:5]

    results = []

    for i, task in enumerate(benchmark, 1):
        print("=" * 80)
        print(f"{i}/{len(benchmark)} — {task['id']}")
        print(task["query"])

        try:
            run = run_dsstar(task)

            correct = False
            if run["execution_success"]:
                correct = evaluate(
                    run["result_json"],
                    task["expected_key"],
                    task["expected"]
                )

        except Exception:
            run = {
                "execution_success": False,
                "returncode": -999,
                "stdout": "",
                "stderr": "",
                "result_path": None,
                "result_json": None
            }
            correct = False

        row = {
            "id": task["id"],
            "dataset": task["dataset"],
            "category": task.get("category", infer_category(task["id"])),
            "query": task["query"],
            "expected_key": task["expected_key"],
            "expected": task["expected"],
            "execution_success": run["execution_success"],
            "correct": correct,
            "result_path": run["result_path"],
            "result_json": run["result_json"],
            "stdout": run["stdout"],
            "stderr": run["stderr"]
        }

        results.append(row)

        print("Execution:", run["execution_success"])
        print("Correct:", correct)

    save_results(results)
    print_summary(results)


def save_results(results):
    ts = datetime.now().strftime("%Y%m%d_%H%M%S")

    json_path = RESULTS_DIR / f"benchmark_results_{ts}.json"
    csv_path = RESULTS_DIR / f"benchmark_results_{ts}.csv"

    with open(json_path, "w", encoding="utf-8") as f:
        json.dump(results, f, indent=2, ensure_ascii=False)

    with open(csv_path, "w", encoding="utf-8", newline="") as f:
        fieldnames = [
            "id", "dataset", "category", "query", "expected_key", "expected",
            "execution_success", "correct", "result_path"
        ]
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()

        for r in results:
            writer.writerow({
                "id": r["id"],
                "dataset": r["dataset"],
                "category": r["category"],
                "query": r["query"],
                "expected_key": r["expected_key"],
                "expected": json.dumps(r["expected"], ensure_ascii=False),
                "execution_success": r["execution_success"],
                "correct": r["correct"],
                "result_path": r["result_path"]
            })

    print("\nSaved:")
    print(json_path)
    print(csv_path)


def print_summary(results):
    total = len(results)
    exec_ok = sum(r["execution_success"] for r in results)
    correct = sum(r["correct"] for r in results)

    print("\n" + "=" * 80)
    print("SUMMARY")
    print("=" * 80)
    print(f"Total tasks: {total}")
    print(f"Execution success: {exec_ok}/{total} = {exec_ok / total:.2%}")
    print(f"Answer accuracy: {correct}/{total} = {correct / total:.2%}")

    print("\nBy category:")
    for cat in sorted(set(r["category"] for r in results)):
        subset = [r for r in results if r["category"] == cat]
        c_total = len(subset)
        c_correct = sum(r["correct"] for r in subset)
        c_exec = sum(r["execution_success"] for r in subset)

        print(
            f"{cat}: "
            f"execution {c_exec}/{c_total} = {c_exec / c_total:.2%}, "
            f"accuracy {c_correct}/{c_total} = {c_correct / c_total:.2%}"
        )


if __name__ == "__main__":
    main()
```

### Benchmark Runner - original dsstar

```python
import json
import re
import csv
import subprocess
from pathlib import Path
from datetime import datetime
import yaml
import uuid

DSSTAR_DIR = Path("/content/DS-Star")
EVAL_DIR = Path("/content/custom_eval")
BENCHMARK_FILE = EVAL_DIR / "benchmark.json"
RESULTS_DIR = EVAL_DIR / "results"
RESULTS_DIR.mkdir(exist_ok=True)

MAX_TASKS = None


def create_temp_config(task_id):
    base_config = DSSTAR_DIR / "config.yaml"

    with open(base_config, "r", encoding="utf-8") as f:
        cfg = yaml.safe_load(f)

    cfg["run_id"] = f"benchmark_{task_id}_{uuid.uuid4().hex[:8]}"
    cfg["auto_debug"] = False
    cfg["debug_attempts"] = 0
    cfg["max_refinement_rounds"] = 1
    cfg["interactive"] = False

    temp_cfg = EVAL_DIR / f"temp_{task_id}_{uuid.uuid4().hex[:8]}.yaml"

    with open(temp_cfg, "w", encoding="utf-8") as f:
        yaml.safe_dump(cfg, f)

    return temp_cfg, cfg["run_id"]


def load_result_json(result_path):
    try:
        if result_path and result_path.exists():
            with open(result_path, "r", encoding="utf-8") as f:
                return json.load(f), None
    except Exception:
        return None, "Could not load result.json"

    return None, "result.json not found"


def find_latest_result_for_run(run_id):
    candidates = sorted(
        DSSTAR_DIR.glob(f"runs/{run_id}/final_output/result.json"),
        key=lambda p: p.stat().st_mtime,
        reverse=True
    )

    if candidates:
        return candidates[0]

    return None


def run_dsstar(task):
    dataset_path = EVAL_DIR / task["dataset"]
    temp_cfg, run_id = create_temp_config(task["id"])

    cmd = [
        "uv", "run", "python", "dsstar.py",
        "--data-files", str(dataset_path),
        "--query",
        task["query"] + f"\nReturn the final answer as valid JSON using exactly this top-level key: {task['expected_key']}.",
        "--max-rounds", "1",
        "--config", str(temp_cfg)
    ]

    stdout = ""
    stderr = ""
    completed = None

    try:
        completed = subprocess.run(
            cmd,
            cwd=DSSTAR_DIR,
            capture_output=True,
            text=True,
            timeout=900
        )

        stdout = completed.stdout
        stderr = completed.stderr

    except subprocess.TimeoutExpired as e:
        stdout = e.stdout or ""
        stderr = e.stderr or ""

        result_path = find_latest_result_for_run(run_id)
        result_json, _ = load_result_json(result_path)

        return {
            "execution_success": result_json is not None,
            "returncode": -1,
            "stdout": stdout,
            "stderr": stderr,
            "result_path": str(result_path) if result_path else None,
            "result_json": result_json
        }

    match = re.search(
        r"OUTPUT:\s*(runs/.+?/final_output/result\.json)",
        stdout
    )

    result_path = None

    if match:
        result_path = DSSTAR_DIR / match.group(1)
    else:
        result_path = find_latest_result_for_run(run_id)

    result_json, _ = load_result_json(result_path)

    return {
        "execution_success": result_json is not None,
        "returncode": completed.returncode if completed else -1,
        "stdout": stdout,
        "stderr": stderr,
        "result_path": str(result_path) if result_path else None,
        "result_json": result_json
    }


def flatten_json(obj):
    return json.dumps(obj, ensure_ascii=False).lower()


def extract_numbers(obj):
    text = flatten_json(obj)
    nums = re.findall(r"-?\d+(?:\.\d+)?(?:e-?\d+)?", text, flags=re.IGNORECASE)
    return [float(x) for x in nums]


def compare_numbers(pred, expected, abs_tol=1e-3, rel_tol=0.01):
    try:
        pred = float(pred)
        expected = float(expected)
    except Exception:
        return False

    if abs(pred - expected) <= abs_tol:
        return True

    if expected != 0:
        return abs(pred - expected) / abs(expected) <= rel_tol

    return False


def fuzzy_numeric_match(obj, expected, abs_tol=1e-3, rel_tol=0.01):
    nums = extract_numbers(obj)

    for n in nums:
        if compare_numbers(n, expected, abs_tol, rel_tol):
            return True

    return False


def get_expected_value(result_json, expected_key):
    if result_json is None:
        return None

    if isinstance(result_json, dict):
        if expected_key in result_json:
            return result_json[expected_key]

        if "final_answer" in result_json:
            final_answer = result_json["final_answer"]

            if isinstance(final_answer, dict) and expected_key in final_answer:
                return final_answer[expected_key]

            return final_answer

    return result_json


def compare_values(pred, expected):
    if isinstance(expected, (int, float)):
        return compare_numbers(pred, expected)

    if isinstance(expected, str):
        return str(expected).lower().strip() in flatten_json(pred)

    if isinstance(expected, list):
        text = flatten_json(pred)
        return all(str(item).lower() in text for item in expected)

    if isinstance(expected, dict):
        text = flatten_json(pred)

        for key, value in expected.items():
            key_ok = str(key).lower() in text

            if isinstance(value, (int, float)):
                value_ok = fuzzy_numeric_match(pred, value)
            else:
                value_ok = str(value).lower() in text

            if not (key_ok and value_ok):
                return False

        return True

    return False


def evaluate(result_json, expected_key, expected):
    if result_json is None:
        return False

    pred = get_expected_value(result_json, expected_key)

    if pred is not None and compare_values(pred, expected):
        return True

    return compare_values(result_json, expected)


def infer_category(task_id):
    if "basic" in task_id:
        return "dataset_understanding"
    if "follow" in task_id:
        return "follow_up"
    if task_id.startswith("wrangle"):
        return "data_wrangling"
    if task_id.startswith("ml"):
        return "machine_learning"
    if task_id.startswith("multi") or task_id.startswith("hard"):
        return "multi_step"
    if task_id.startswith("stat"):
        return "statistical_analysis"
    return "other"


def validate_benchmark(benchmark):
    errors = []

    for task in benchmark:
        if "id" not in task:
            errors.append("Task missing id")
            continue

        if "dataset" not in task:
            errors.append(f"{task['id']}: missing dataset")

        if "query" not in task:
            errors.append(f"{task['id']}: missing query")

        if "expected" not in task:
            errors.append(f"{task['id']}: missing expected")

        if "expected_key" not in task:
            errors.append(f"{task['id']}: missing expected_key")

        if "dataset" in task:
            dataset_path = EVAL_DIR / task["dataset"]
            if not dataset_path.exists():
                errors.append(f"{task['id']}: missing dataset {dataset_path}")

    if errors:
        raise ValueError("\n".join(errors))


def main():
    with open(BENCHMARK_FILE, "r", encoding="utf-8") as f:
        benchmark = json.load(f)

    validate_benchmark(benchmark)

    if MAX_TASKS is not None:
        benchmark = benchmark[:MAX_TASKS]

    results = []

    for i, task in enumerate(benchmark, 1):
        print("=" * 80)
        print(f"{i}/{len(benchmark)} — {task['id']}")
        print(task["query"])

        try:
            run = run_dsstar(task)

            correct = False
            if run["execution_success"]:
                correct = evaluate(
                    run["result_json"],
                    task["expected_key"],
                    task["expected"]
                )

        except Exception:
            run = {
                "execution_success": False,
                "returncode": -999,
                "stdout": "",
                "stderr": "",
                "result_path": None,
                "result_json": None
            }
            correct = False

        row = {
            "id": task["id"],
            "dataset": task["dataset"],
            "category": task.get("category", infer_category(task["id"])),
            "query": task["query"],
            "expected_key": task["expected_key"],
            "expected": task["expected"],
            "execution_success": run["execution_success"],
            "correct": correct,
            "returncode": run["returncode"],
            "result_path": run["result_path"],
            "result_json": run["result_json"],
            "stdout": run["stdout"],
            "stderr": run["stderr"]
        }

        results.append(row)

        print("Execution:", run["execution_success"])
        print("Correct:", correct)

    save_results(results)
    print_summary(results)


def save_results(results):
    ts = datetime.now().strftime("%Y%m%d_%H%M%S")

    json_path = RESULTS_DIR / f"benchmark_results_{ts}.json"
    csv_path = RESULTS_DIR / f"benchmark_results_{ts}.csv"

    with open(json_path, "w", encoding="utf-8") as f:
        json.dump(results, f, indent=2, ensure_ascii=False)

    with open(csv_path, "w", encoding="utf-8", newline="") as f:
        fieldnames = [
            "id", "dataset", "category", "query", "expected_key", "expected",
            "execution_success", "correct", "returncode", "result_path"
        ]

        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()

        for r in results:
            writer.writerow({
                "id": r["id"],
                "dataset": r["dataset"],
                "category": r["category"],
                "query": r["query"],
                "expected_key": r["expected_key"],
                "expected": json.dumps(r["expected"], ensure_ascii=False),
                "execution_success": r["execution_success"],
                "correct": r["correct"],
                "returncode": r["returncode"],
                "result_path": r["result_path"]
            })

    print("\nSaved:")
    print(json_path)
    print(csv_path)


def print_summary(results):
    total = len(results)
    exec_ok = sum(r["execution_success"] for r in results)
    correct = sum(r["correct"] for r in results)

    print("\n" + "=" * 80)
    print("SUMMARY")
    print("=" * 80)
    print(f"Total tasks: {total}")
    print(f"Execution success: {exec_ok}/{total} = {exec_ok / total:.2%}")
    print(f"Answer accuracy: {correct}/{total} = {correct / total:.2%}")

    print("\nBy category:")
    for cat in sorted(set(r["category"] for r in results)):
        subset = [r for r in results if r["category"] == cat]
        c_total = len(subset)
        c_correct = sum(r["correct"] for r in subset)
        c_exec = sum(r["execution_success"] for r in subset)

        print(
            f"{cat}: "
            f"execution {c_exec}/{c_total} = {c_exec / c_total:.2%}, "
            f"accuracy {c_correct}/{c_total} = {c_correct / c_total:.2%}"
        )


if __name__ == "__main__":
    main()
```
